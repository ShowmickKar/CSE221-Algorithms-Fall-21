```python
import time
from matplotlib import pyplot as plt
from maximum_subarray_problem import maximum_subarray_bruteforce, kadens_algorithm
from maximum_subarray_problem import array_container
```

```python
time_brute_force = []
time_kadens_algorithm = []
```

Now running both algorithms and recording their runtime....

```python
for a in array_container:
    start = time.time()
    maximum_subarray_bruteforce(a)
    end = time.time()
    time_brute_force.append(end - start)
    start = time.time()
    kadens_algorithm(a)
    end = time.time()
    time_kadens_algorithm.append(end - start)
```

Runtime graph of a Naive Brute force Solution

```python
plt.plot(time_brute_force)
plt.show()
```

Runtime graph for kaden's algorithm

```python
plt.plot(time_kadens_algorithm)
plt.show()
```
